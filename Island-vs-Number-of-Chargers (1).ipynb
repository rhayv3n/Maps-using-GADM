```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Please make sure to replace it with your actual dataset.
data = {
    'Island': ['Oahu', 'Maui', 'Oahu', 'Kauai', 'Maui', 'Hawaii', 'Oahu', 'Hawaii', 'Kauai'],
    'Number of Chargers': [3, 4, 2, 1, 5, 4, 3, 2, 1]
}
CDS = pd.DataFrame(data)

# Define color mapping for the islands
color_mapping = {
    'Oahu': 'skyblue',
    'Maui': 'lightgreen',
    'Kauai': 'coral',
    'Hawaii': 'gold'
}

# Group the data by 'Island' and sum the 'Number of Chargers'
chargers_per_island = CDS.groupby('Island')['Number of Chargers'].sum().reset_index()

# Calculate the percentage of chargers for each island
total_chargers = chargers_per_island['Number of Chargers'].sum()
chargers_per_island['Percentage'] = (chargers_per_island['Number of Chargers'] / total_chargers) * 100

# Print out the percentage values
print(chargers_per_island[['Island', 'Percentage']])

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Create a bar plot
plt.figure(figsize=(10, 6))
barplot = sns.barplot(
    x='Island', 
    y='Number of Chargers', 
    data=chargers_per_island, 
    palette=color_mapping
)

# Annotate the percentage on top of each bar
for index, row in chargers_per_island.iterrows():
    barplot.text(
        row.name, 
        row['Number of Chargers'], 
        f"{row['Percentage']:.2f}%", 
        color='black', 
        ha="center"
    )

# Set plot titles and labels
plt.title('Percentage of EV Chargers Per Island in Hawaii')
plt.xlabel('Island')
plt.ylabel('Total Number of Chargers')

# Show the plot
plt.show()
```

Key Points:
- Maui leads with 36% of Hawaii's total EV chargers, followed closely by Oahu at 32%.

- Kauai has the fewest chargers, making up just 8% of the total, indicating potential areas for infrastructure development.

Explanation: 
The distribution of electric vehicle chargers across the Hawaiian Islands shows a concentration in Maui and Oahu, which together account for 68% of the chargers. Hawaii Island, while significant in landmass, holds 24% of the state's charging infrastructure, reflecting a moderate charger density.
